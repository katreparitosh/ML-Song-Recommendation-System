```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
df = pd.read_csv('A:\\Data Analysis Jupyter\\Song Recommeder\\Song Data.csv')
df.head()
```

# Using Cosine Similarity 
Cosine Similarity is a measure to find similarity between two file/docs.
        
cos(#) = (file1)(file2) / |file|^0.5 * |file2|^0.5, *0.5 is square root*

If angle is small(the files share a lot in common), i.e the cosine is large.
If angle is large(the files are not so similar), i.e the cosine  is small.

```python
data_similarity = cosine_similarity(songs.T)
data_similarity
```

data_similarity matrix is obtained

```python
type(data_similarity) # is an ndarray = n multidimentional array
```

Converting ndarra to DataFrame

```python
data_similarity_df = pd.DataFrame(data_similarity, columns = (songs.columns), index = (songs.columns))
data_similarity_df.head() 
```

This the cosine_similarity matrix.

```python
data_similarity_df.index.is_unique # check if no repeated songs are present
```

Apply NearestNeighbors Algorithm.
Supply the data_similarity_df matrix to get recommendations.

```python
neigh = NearestNeighbors(n_neighbors = 285)
neigh.fit(data_similarity_df) # fit the data
```

```python
# copy the predicted data to a new dataframe model
# kneighbors() - finds k-neighbors of a point. Returns indices of and distances to the neighbors of each point.

model = pd.DataFrame(neigh.kneighbors(data_similarity_df, return_distance = False))
model.head() # gives integer values instead of names
```

```python
final_model = pd.DataFrame(data_similarity_df.columns[model], index = data_similarity_df.index)
final_model
# gives names with respect to songs
```

```python
final_model.head()
```

Above model gives all recommendations

```python
top10 = final_model[list(final_model.columns[:11])]
top10.head()
```
